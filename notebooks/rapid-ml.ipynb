```python
import numpy as np
import statsmodels.api as smf
import pandas as pd
```

```python
df = pd.read_csv('../data/fake_fraud_data.csv',index_col=0)
df
```

```python

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, random_state=42, test_size=0.1)
train
```

```python
train.to_csv('../data/training.csv')
test.to_csv('../data/testing.csv')
```

```python
train.corr()
```

```python
train[['age','years_in_labour_force']].corr().iloc[0:1]
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

mod = smf.ols(formula='years_in_labour_force ~ age', data=train)
res = mod.fit()
res.summary()
```

```python
res.summary().tables[0]
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

mod = smf.ols(formula='committed ~ age + years_in_labour_force + years_of_further_education+ miles_resided_from_coast', data=train)
res = mod.fit()
res.summary()
```
